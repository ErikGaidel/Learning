```python
with open("input.txt", "r") as inp_file:
    t = int(inp_file.readline())
    for _ in range(t):
        n, k = map(int, inp_file.readline().split())
        nk = n * k
        h_dish = []
        n_dish = []

        for i in range(n):
          m, p =  map(float, inp_file.readline().split())
          h_dish.append((p, m, m*p))
        h_dish.sort(reverse=True)
        
        s_n = 0
        for i in range(k):
          d, q =  map(float, inp_file.readline().split())
          dq = d * q
          s_n += dq
          n_dish.append((q, d, dq))
        n_dish.sort(reverse=True)
        

        C = 0

        vspom = []
        for i in range(len(h_dish)):
            for j in range(1,i+1):
              C += h_dish[j][0]
            if C/(i+1) == n_dish[0][0]:
              for j in range(1,i+1):
                vs = [(i+1)*[0]]
                for k in range(1,h_dish[j][1]+1):
                  vs[k] += 1
                vspom.appen(vs)
        for v in vspom:
          if v.sum() == n_dish[0][1]:
            print(*v)
```
